```python
import pandas as pd
import numpy as np

# Función para crear un DataFrame ficticio con un número especificado de filass
def crear_dataframe_ficticio(n_filas):
    # Crear datos ficticios
    data = {
        'Movement Number': np.random.randint(10000000, 99999999, size=n_filas),
        'Order Number': np.random.randint(1000000, 9999999, size=n_filas),
        'Event': np.random.choice(['XDL', 'DLT', 'HCT', 'TRP'], size=n_filas),
        'Company ID': np.random.choice(['LOBREG', 'BISREG03', 'NATREG01'], size=n_filas),
        'City': np.random.choice(['REGINA,SK/', 'CALGARY,AB/', 'TORONTO,ON/'], size=n_filas),
        'Zipcode': np.random.choice(['S4P 3W5', 'S4K 0A2', 'S4N 5W3', 'T2P 1N6', 'M5V 3K4'], size=n_filas),
        'Earliest Appt': pd.date_range(start='2024-01-01', periods=n_filas, freq='H'),
        'Latest Appt': pd.date_range(start='2024-01-01', periods=n_filas, freq='H'),
        'Net Quantity Ordered': np.random.randint(0, 100, size=n_filas),
        'Ordered Volume': np.random.randint(0, 1000, size=n_filas),
        'Ordered Weight': np.random.randint(0, 5000, size=n_filas),
    }

    # Crear el DataFrame
    df_ficticio = pd.DataFrame(data)

    return df_ficticio

# Ejemplo: crear un DataFrame con 10 filas
df_ficticio = crear_dataframe_ficticio(20)
df_ficticio
```
